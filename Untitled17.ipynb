```python
from pylatex import Document, Section, Subsection, Command, Tabular
from pylatex.utils import bold
```

```python
#Creamos un documento
doc = Document()
```

```python
#Ponemos un titulo al doc
doc.preamble.append(Command('Titulo', 'Documento creado con Python'))
doc.preamble.append(Command('Autor', 'Nombre'))
doc.preamble.append(Command('fecha', '14 de abril 2023'))
doc.append(Command('maketitle'))
```

```python
#Añade una sección al documento
with doc.create(Section('Seccion 1')):
    doc.append('Texto de la seccion 1')
    
    #Añade una subseccion
    
    with doc.create(Subsection('Subseccion 1.1')):
        doc.append('Texto de la subseccion')
#Se pueden añadir cuantas secciones y subseccione quieran 
```

```python
#Añade una tabla
with doc.create(Tabular('ccc')) as table:
    table.add_hline()
    table.add_row((bold('Columna 1'), bold('Columna 2'), bold('Columna 3')))
    table.add_hline()
    table.add_row(('1', '2', '3'))
    table.add_row(('4', '5', '6'))
    table.add_hline()
```

```python
#Genera el documento 
doc.generate_pdf('ejemplo',clean_tex=False)
```
